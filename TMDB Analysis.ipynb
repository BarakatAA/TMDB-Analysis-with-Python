### Exploratory data analysis


Question 1: Will the popularity rating of a movie of a movie affect its revenue

```python
sns.regplot(x = 'popularity', y = 'revenue_adj', data = df)
plt.title ('Popularity level vs Revenue Generation')

# The plot shows a positive correlation and that a high popularity level leads to more revenue generation as the more popular a movie, the higher revenue it generates
```

Question 2

Can the budget allocated to a movie affect its revenue generation?

```python
sns.regplot(x = 'revenue_adj', y = 'budget_adj', data = df)
plt.title ('Budget Allocation vs Revenue Generation')
# Shows a positive correlation too and that movies with higher budgets gave more revenue 
```

Question 3


What movies are the most popular?

```python
most_popular_movies = df.sort_values(by = 'popularity', ascending = False).head(10)

sns.barplot(x = 'original_title', y = 'popularity', data = most_popular_movies)
plt.xticks(rotation = 90)
plt.title = 'Most Popular movies'

plt.show()
```

The movie with the most popularity is Jurassic World followed by Mad Max Fury and Interstellar

### Question 4
